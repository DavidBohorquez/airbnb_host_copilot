# M2 — Embeddings & chunking

**Goal:** turn text into vectors with a multilingual model, and understand chunking.

**Skills:** what an embedding is, tokenization vs chunking, multilingual models, e5 prefixes, batching,
caching vectors to disk.

## Safe checkpoint

- Encode 20 listing documents → shape `(20, 384)` for e5-small.
- `cosine(encode_query("appartement calme à Chartrons"), <a quiet-Chartrons passage>)` is clearly higher
  than against a random noisy-listing passage.
- Vectors are unit-norm: `np.linalg.norm(v) ≈ 1`.

````python
from host_copilot.data.loaders import load_listings
from host_copilot.documents import listing_to_document
from host_copilot.embeddings import Embedder
import numpy as np

df = load_listings()
docs = [listing_to_document(df.iloc[i]) for i in range(20)]
emb = Embedder()
V = emb.encode_passages(docs)
print('shape', V.shape)
print('norm_doc0', round(float(np.linalg.norm(V[0])), 4))
q = emb.encode_query('appartement calme a Chartrons')
sims = V @ q
print('best_sim', round(float(sims.max()), 3), 'worst_sim', round(float(sims.min()), 3))
````

# M3 — Vector search from scratch (NumPy)

**Goal:** implement exact top-k similarity search yourself, before any library. This is the single most
important conceptual milestone.

**Skills:** cosine vs dot vs Euclidean (L2), exact/brute-force search, why it doesn't scale.

## Experiment (do this, it builds intuition)

- Run the same query with cosine (dot of normalized) and with raw L2 distance. Confirm the **ranking** is
  the same but scores differ.
- Time a search over all listings. Note it in your notebook — you'll compare against FAISS in M4.

## Safe checkpoint

- Query "appartement lumineux proche tramway" returns plausibly matching listings in the top 5.
- Self-query (a document's own vector) returns that document as rank 1 with score ≈ 1.0.

````python
import time, numpy as np
from host_copilot.config import get_settings
from host_copilot.data.loaders import load_listings
from host_copilot.documents import listing_to_document
from host_copilot.embeddings import Embedder
from host_copilot.vectorstore.numpy_store import NumpyStore

df = load_listings()
docs = [listing_to_document(df.iloc[i]) for i in range(len(df))]
ids = df['id'].tolist()
names = dict(zip(ids, df['name']))
cache = get_settings().artifacts_path / 'listings.npy'
V = Embedder().encode_passages(docs, cache_path=cache)
store = NumpyStore(V, ids)
top = store.search(V[0], k=1)[0]
print('self_query rank1 id==ids0:', top[0] == ids[0], 'score', round(top[1],4))

from host_copilot.embeddings import Embedder as E
q = E().encode_query('appartement lumineux proche tramway')
cos = store.search(q, k=5)
print('TOP5 cosine:')
for i,(lid,s) in enumerate(cos): print(' ', i+1, round(s,3), str(names[lid])[:60])

l2 = store.search_l2(q, k=5)
print('ranking cosine==l2:', [c[0] for c in cos] == [d[0] for d in l2])

t=time.perf_counter()
for _ in range(100): store.search(q, k=8)
print('avg_search_ms', round((time.perf_counter()-t)/100*1000,3), 'over', len(ids), 'vecs')
````

# M4 — FAISS (approximate nearest neighbour)

**Goal:** index the same vectors in FAISS, verify it matches your NumPy result on a flat index, then try
an approximate index and measure the speed/recall tradeoff.

**Skills:** ANN, `IndexFlatIP` vs `IndexIVFFlat`/`IndexHNSWFlat`, recall vs speed, index persistence.

## Experiment

- Compare FlatIP top-k vs your NumPy top-k → should be **identical**.
- Compare IVF (low `nprobe`) vs FlatIP → measure recall@10 (overlap of result sets) and the speed gain.
- Time queries vs M3. Record the numbers.

## Safe checkpoint

- FlatIP results == NumPy results on the same query.
- You can state your IVF recall@10 and speedup vs flat.
- Index round-trips through `save`/`load` and still returns the same results.

````python
import time, numpy as np
from host_copilot.config import get_settings
from host_copilot.data.loaders import load_listings
from host_copilot.embeddings import Embedder
from host_copilot.vectorstore.numpy_store import NumpyStore
from host_copilot.vectorstore.faiss_store import FaissStore

df = load_listings(); ids = df['id'].tolist()
V = np.load(get_settings().artifacts_path / 'listings.npy').astype('float32')
q = Embedder().encode_query('appartement lumineux proche tramway')

npy = NumpyStore(V, ids)
flat = FaissStore(V.shape[1]); flat.add(V, ids)
ivf = FaissStore(V.shape[1], index_type='ivf', nlist=100); ivf.add(V, ids); ivf.nprobe = 8

np_top = [i for i,_ in npy.search(q, 10)]
flat_top = [i for i,_ in flat.search(q, 10)]
ivf_top = [i for i,_ in ivf.search(q, 10)]
print('flat==numpy:', flat_top == np_top)
print('ivf recall@10:', len(set(ivf_top) & set(flat_top)) / 10)

def t(store):
    s=time.perf_counter()
    for _ in range(200): store.search(q, 10)
    return round((time.perf_counter()-s)/200*1000, 4)
print('ms numpy/flat/ivf:', t(npy), t(flat), t(ivf))

p = str(get_settings().artifacts_path / 'flat.faiss')
flat.save(p); rt = FaissStore.load(p)
print('save/load equal:', [i for i,_ in rt.search(q,10)] == flat_top)
````

**Goal:** move to a real vector database. Persist all listing vectors once, and query with a **metadata

**Skills:** production vector DB, collections, upsert, persistence, `where=` metadata filtering.

## Steps

## Safe checkpoint

- `make ingest` completes; `artifacts/chroma/` exists and survives a restart (re-opening the client finds
  the data — no re-ingest).
- A query with `where={"peer_key": "<some real key>"}` returns **only** listings from that peer group.
- Count in the collection ≈ number of listings ingested.

````python
from host_copilot.ingest import main
main()

import numpy as np
from host_copilot.data.loaders import load_listings
from host_copilot.embeddings import Embedder
from host_copilot.vectorstore.chroma_store import ChromaStore

df = load_listings()
pk = df['peer_key'].value_counts().index[0]   # a populated peer group
print('test peer_key:', pk, 'count_in_df:', int((df['peer_key']==pk).sum()))

# fresh client -> proves persistence (no re-ingest)
store = ChromaStore()
print('collection count:', store.count(), 'df rows:', len(df))

q = Embedder().encode_query('appartement calme et lumineux')
res = store.query(q, k=5, where={'peer_key': pk})
print('filtered results all in peer group:', all(r['metadata']['peer_key']==pk for r in res))
print('sample:', [(r['metadata']['room_type'], round(r['score'],3)) for r in res][:3])
````

# M6 — Retrieval logic

**Goal:** the retrieval layer Host Copilot actually calls: given a `listing_id`, return its valid comps; given a free-text need, run hybrid (semantic + keyword) search.

**Skills:** peer-group filtering, top-k comp selection, hybrid search, the idea of reranking.

## Experiment

Pick a query with a specific term (e.g. "parking gratuit"). Compare pure-semantic vs hybrid top-5 — hybrid
should surface the listings that literally mention it.

## Safe checkpoint

- `get_comps(<id>)` returns only same-peer-group listings, never the listing itself.
- On a keyword-specific query, hybrid ranks the literal-match listing above pure semantic does.


````python
from host_copilot.data.loaders import load_listings
from host_copilot.retrieval import get_comps, hybrid_search, _store, _embedder

df = load_listings()
lid = int(df['id'].iloc[0])
pk0 = df.set_index('id').loc[lid, 'peer_key']
comps = get_comps(lid, k=5)
print('comps all same peer_key:', all(c['metadata']['peer_key'] == pk0 for c in comps))
print('self excluded:', all(str(c['id']) != str(lid) for c in comps))
print('n comps:', len(comps))

pk = df['peer_key'].value_counts().index[0]
store = _store()
q = 'parking'
sem = [r['id'] for r in store.query(_embedder().encode_query(q), k=5, where={'peer_key': pk})]
hyb = [r['id'] for r in hybrid_search(q, pk, k=5)]
g = store.get_group(pk); docs = dict(zip(g['ids'], g['documents']))
lit = lambda ids: [('parking' in docs[i].lower()) for i in ids]
print('semantic literal-parking hits:', lit(sem))
print('hybrid   literal-parking hits:', lit(hyb))
````

````python
q = "rooftop jacuzzi"
term = "jacuzzi"          # the rare literal token to look for

g = store.get_group(pk)
docs = dict(zip(g['ids'], g['documents']))
n_hit = sum(term in d.lower() for d in g['documents'])
print(f"docs in pk mentioning '{term}': {n_hit}/{len(g['ids'])}")

sem = [r['id'] for r in store.query(_embedder().encode_query(q), k=5, where={'peer_key': pk})]
hyb = [r['id'] for r in hybrid_search(q, pk, k=5)]
hit = lambda ids: [term in docs[i].lower() for i in ids]
print("semantic:", hit(sem))
print("hybrid:  ", hit(hyb))
````
